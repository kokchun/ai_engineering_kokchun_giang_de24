# Validation without pydantic

```python
# Student class
# name, age, gender

# instantiate instance from Student


class Student:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    # "dunder repper" -> representation of this instance
    def __repr__(self):
        return (
            f"Student(name='{self.name}', age={self.age}, gender='{self.gender}')"
        )


s1 = Student("Kokchun", 34, "M")
s1
```
