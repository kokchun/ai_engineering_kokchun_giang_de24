validation

```python

class Student:
    def __init__(self, name, age, gender):
        self.name = name # attribute
        self.age = age # property
        self.gender = gender # attribute

    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, age):
        # validation code here
        if not isinstance(age, (int, float)):
            raise TypeError(f"age must be number not {type(age)}")

        if not 0<=age<125:
            raise ValueError(f"Age must be between 0 and 124, not {age}")

        self._age = age

    # special method
    # "dunder repper" -> representation of this instance
    def __repr__(self):
        return (
            f"Student(name='{self.name}', age={self.age}, gender='{self.gender}')"
        )
try:
    Student("Adam", "femtioåtta", "M")
except TypeError as err:
    print(err)

try:
    Student("Adam", -5, "M")
except ValueError as err:
    print(err)

s3 = Student("Adam", 124, "M")
s3
```
