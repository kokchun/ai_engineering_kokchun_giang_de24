## pdf to text

```python
from pypdf import PdfReader

reader = PdfReader("data/Netherland_Dwarf_rabbit.pdf")
reader
```

```python
all_text = ""

for page in reader.pages:
    all_text += page.extract_text() + "\n"

len(all_text)
```
